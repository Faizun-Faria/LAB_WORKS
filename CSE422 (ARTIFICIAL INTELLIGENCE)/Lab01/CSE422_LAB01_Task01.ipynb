```
from queue import Queue

def append_value(dict_obj, key, value):
  if key in dict_obj:
    if not isinstance(dict_obj[key], list):
      dict_obj[key] = [dict_obj[key]]
    dict_obj[key].append(value)
  else:
    dict_obj[key] = value

url = '/content/sample_data/Task1.txt'
file1 = open(url, 'r') 
lines = file1.read().split("\n")
count = 0
nodes = int(lines[count])
count+=1
edges = int(lines[count])
count+=1
color = {}
depth = {}
graph = {}
for i in range(0,edges):
  color[i] = 0
  depth[i] = 0
for i in range(0,edges):
  x, y = lines[count].split()
  count+=1
  u = int(x)
  v = int(y) 
  append_value(graph, x, v)
  append_value(graph, y, u)
destination = int(lines[count])
q = Queue()
q.put(str(0))
color[0] = 1
flag = False
while True:
  top = q.get()
  adj_list = graph[top]
  for i in adj_list:
    if color[i] == 0:
      depth[i] = depth[int(top)] + 1
      if i == destination and flag == False:
        flag = True
        print(depth[i])
        break
      color[i] = 1
      q.put(str(i))
  if q.empty() or flag:
    break  
graph.clear()
```
