```
from queue import Queue
def append_value(dict_obj, key, value):
  if key in dict_obj:
    if not isinstance(dict_obj[key], list):
      dict_obj[key] = [dict_obj[key]]
    dict_obj[key].append(value)
  else:
    dict_obj[key] = value

url = '/content/sample_data/Task3.txt'
file1 = open(url, 'r') 
lines = file1.read().split("\n")
count = 0
nodes = int(lines[count])
count+=1
edges = int(lines[count])
count+=1
graph = {}
graph.clear()
color = {}
depth = {}
pos = {}
for i in range(0,edges):
  color[i] = 0
  depth[i] = 0
  pos[i] = 0
for i in range(0,edges):
  x, y = lines[count].split()
  count += 1
  u = int(x)
  v = int(y) 
  append_value(graph, x, v)
  append_value(graph, y, u)
victim = int(lines[count])
count+=1
numOfAttackers= int(lines[count])
count+=1
for i in range(0, numOfAttackers):
  num = int(lines[count])
  count+=1
  pos[num] = 1
q = Queue()
q.put(str(victim))
color[victim] = 1
flag = False
while True:
  top = q.get()
  adj_list = graph[top]
  for i in adj_list:
    if color[i] == 0:
      depth[i] = depth[int(top)] + 1
      if pos[i] == 1:
        print(depth[i])
        flag = True
        break
      color[i] = 1
      q.put(str(i))
  if q.empty() or flag==True:
    break     
graph.clear()
```
