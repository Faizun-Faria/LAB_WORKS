```
import random
import numpy as np

numberOfTurns = int(input())
treeDepth = 2*numberOfTurns
branchingFactor = int(input())
leafNodes = branchingFactor ** treeDepth
startRange = int(input())
endRange = int(input())
print( "Depth:" + str(treeDepth) )
print( "Branch:" + str(branchingFactor) )
print( "Terminal States (Leaf Nodes): " + str(leafNodes) )
nonLeaf = 0
for i in range(treeDepth):
  nonLeaf +=  (branchingFactor ** i) 
values = [0]*(nonLeaf+leafNodes)
child = []
cnt = 1
for i in range(nonLeaf):
    column = []
    for j in range(branchingFactor):
        column.append(cnt)
        cnt += 1
    child.append(column)   
for i in range(nonLeaf,(nonLeaf+leafNodes)):
  randomNumber = random.randint(startRange, endRange)
  values[i] = randomNumber
topNode = 0
comparisonsAfterABP = 0
def minimax(position, depth, alpha, beta , maximizingPlayer):
  global comparisonsAfterABP
  if depth == 0:    
    comparisonsAfterABP += 1
    return values[position]
  if maximizingPlayer:
    maxEval = -np.inf
    for i in range(branchingFactor):
      eval = minimax(child[position][i], depth-1, alpha, beta, False)
      maxEval = max(maxEval, eval)
      alpha = max(alpha, eval)
      if beta <= alpha:
        break    
    return maxEval
  else:
    minEval = +np.inf
    for i in range(branchingFactor):
      eval = minimax(child[position][i], depth-1, alpha, beta, True)
      minEval = min(minEval, eval)
      beta = min(beta, eval)
      if beta <= alpha:
        break    
    return minEval 
minimax (topNode, treeDepth, -np.inf, +np.inf, True)
print("Maximum amount: "+str(branchingFactor))
print("Comparisons: "+str(leafNodes))
print("Comparisons: "+str(comparisonsAfterABP))
```
