There are some values of 'ca' which are out of range. So deleting them from dataframe.

```
dataframe = dataframe.drop((dataframe[dataframe['ca'] > 3]).index, axis = 0)
dataframe.shape
```

##Feature engineering. Introducing new column named angina which contains yes and no.

```
angina = []
cp = dataframe['cp'].to_list()
for i in range(len(dataframe)): 
  angina.append("yes" if(cp[i]==1 or cp[i]==2) else "no")
dataframe['angina'] = angina
dataframe.head(10)
```

##Scaling all the values between 0-1 with proper scaling technique(MinMax)

```
run_columns = ['cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
```

```
print("per-feature minimum before scaling:\n {}".format(
    dataframe[run_columns].min(axis=0)))
print("per-feature maximum before scaling:\n {}".format(
    dataframe[run_columns].max(axis=0)))
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dataframe[run_columns] = scaler.fit_transform(dataframe[run_columns])
```

```
print("per-feature minimum after scaling:\n {}".format(
    dataframe[run_columns].min(axis=0)))
print("per-feature maximum after scaling:\n {}".format(
    dataframe[run_columns].max(axis=0)))
```
